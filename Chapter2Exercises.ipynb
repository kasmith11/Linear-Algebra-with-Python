```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#create a 2D vector named v
v = np.array([1,2])
for i in range(10):
    scalar = np.random.randn()
    multiple = v*scalar
    plt.plot([0, multiple[0]], [0, multiple[1]])

# All of the vectors are on the same line because they were "sclaed" by the random scalar
```
